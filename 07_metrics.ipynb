```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
output_path = "best_model/"

files = os.listdir(output_path)
accuracy = [file for file in files if "Accuracy" in file]
loss = [file for file in files if "Loss" in file]
f1 = [file for file in files if "F1" in file]
recall = [file for file in files if "Recall" in file]
```

```python
def plot_graphs(files, title):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    for file in files:
        df = pd.read_csv(output_path + file)
        if "test" in file:
            if "100" in file:
                ax[0].plot(df["Step"], df["Value"], label= file.split("_")[7] + " test")
            else:
                ax[1].plot(df["Step"], df["Value"], label= file.split("_")[7] + " test")
        else:
            if "100" in file:
                ax[0].plot(df["Step"], df["Value"], label= file.split("_")[7] + " train")
            else:
                ax[1].plot(df["Step"], df["Value"], label= file.split("_")[7] + " train")
    ax[0].set_title(title + " 100 epochs")
    ax[1].set_title(title + " 150 epochs")
    ax[0].legend()
    ax[1].legend()
    ax[0].set_xlabel("Epoch")
    ax[1].set_xlabel("Epoch")
    ax[0].set_ylabel(title)
    ax[1].set_ylabel(title)
    ax[0].grid()
    ax[1].grid()
    plt.show()
```

```python
plot_graphs(accuracy, "Accuracy")
```

```python
plot_graphs(loss, "Loss")
```

```python
plot_graphs(f1, "F1_weighted")
```

```python
plot_graphs(recall, "Recall_weighted")
```
